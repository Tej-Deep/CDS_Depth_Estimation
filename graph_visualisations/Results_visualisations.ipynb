```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dvalues_nyuandkittitrained_testkitti_data = pd.read_csv('dvalues_nyuandkittitrained_testkitti.csv')
dvalues_nyuandkittitrained_testnyu_data = pd.read_csv('dvalues_nyuandkittitrained_testnyu.csv')
dvalues_nyutrained_testkitti_data = pd.read_csv('dvalues_nyutrained_testkitti.csv')
dvalues_nyutrained_testnyu_data = pd.read_csv('dvalues_nyutrained_testnyu.csv')
silogloss_nyuandkittitrained_testkitti_data = pd.read_csv('silogloss_nyuandkittitrained_testkitti.csv')
silogloss_nyuandkittitrained_testnyu_data = pd.read_csv('silogloss_nyuandkittitrained_testnyu.csv')
silogloss_nyutrained_testkitti_data = pd.read_csv('silogloss_nyutrained_testkitti.csv')
silogloss_nyutrained_testnyu_data = pd.read_csv('silogloss_nyutrained_testnyu.csv')
```

```python
def plot_dvalues_nyuandkittitrained_testkitti():
    df = pd.DataFrame(dvalues_nyuandkittitrained_testkitti_data)
    df = df[df['Model']!='A']
    my_color = ['gold', 'powderblue','plum']
    ax = df.plot.bar(x='Model',color=my_color, figsize=(10,9))
    ax.set_xlabel('Models')
    ax.set_ylabel('d1, d2 and d3 values')
    for container in ax.containers:
        ax.bar_label(container)
    plt.title("d values - models trained on: NYU and KITTI dataset, tested on: KITTI dataset")
    plt.xticks(rotation = 0)
    plt.tight_layout()
    plt.show()
    
plot_dvalues_nyuandkittitrained_testkitti()
```

```python
def plot_dvalues_nyuandkittitrained_testnyu():
    df = pd.DataFrame(dvalues_nyuandkittitrained_testnyu_data)
    df = df[df['Model']!='A']
    my_color = ['gold', 'powderblue','plum']
    ax = df.plot.bar(x='Model',color=my_color, figsize=(10,9))
    ax.set_xlabel('Models')
    ax.set_ylabel('d1, d2 and d3 values')
    for container in ax.containers:
        ax.bar_label(container)
    plt.title("d values - models trained on: NYU and KITTI dataset, tested on: NYU dataset")
    plt.xticks(rotation = 0)
    plt.tight_layout()
    plt.show()
    
plot_dvalues_nyuandkittitrained_testnyu()
```

```python
def plot_silogloss_nyuandkittitrained_testkitti():
    df = pd.DataFrame(silogloss_nyuandkittitrained_testkitti_data)
    df = df[df['Model']!='A']
    X = list(df.iloc[:, 0])
    Y = list(df.iloc[:, 1])
    plt.bar(X, Y, color="darkseagreen")
    plt.title("Silog Loss - models trained on: NYU and KITTI dataset, tested on: KITTI dataset")
    plt.xlabel("Models")
    plt.ylabel("Silog Loss")
    plt.show()
    
plot_silogloss_nyuandkittitrained_testkitti()
```
